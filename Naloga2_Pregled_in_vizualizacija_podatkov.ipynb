```
df_rank= pd.read_csv("university_rank_2020.csv",index_col="University", decimal=",")
```

```
df_other = pd.read_csv("university_rank_other.csv", sep=" ",index_col=0, decimal=".")
```

```
df_student= pd.read_excel("university_rank_students.xlsx",index_col="University", sheet_name=0)
```

```
df = pd.concat([df_other, df_rank, df_student], axis=1)
```

```
#3. Izpišite število vrstic in stolpcev,
#ki jih ima novo ustvarjeni df in preglejte kako izgleda prvih 10 vrstic.
df.shape
```

```
df.head(10)
```

```
#4. Preverite koliko je manjkajočih podatkov v posameznem stolpcu.
df.isnull().sum()
```

```
#5. Manjkajoče vrednosti za stolpec International_Outlook ročno dopolnite glede na stolpec Teaching -
#če je vrednost v posamezni vrstici za atribut Teaching:
#< 20 - dopolnite stolpec International_Outlook z vrednostjo "bad"
#> 60  - dopolnite stolpec International_Outlook z vrednostjo "excellent"
#vmesne vrednosti pa dopolnite z vrednostjo "good"

df.loc[df['Teaching'] < 20, 'International_Outlook'] = 'bad'
df.loc[df['Teaching'] > 60, 'International_Outlook'] = 'excellent'
df.loc[(df['Teaching'] >= 20) & (df['Teaching'] <= 60), 'International_Outlook'] = 'good'
```

```
#6. Manjkajoče vrednosti v ostalih stolpcih dopolnite s povprečjem stolpca.
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```
#7. Izpišite koliko je univerz v posamezni državi.
uni_drz = df['Country'].value_counts()
print(uni_drz)
```

```
# 8. Izpišite vse univerze, ki se nahajajo v Sloveniji
slo_uni = df[df['Country'] == 'Slovenia']
print(slo_uni)
```

```
# 9. Ustvarite nov stolpec Staff Skills
df['Staff Skills'] = (df['Teaching'] + df['Research']) / 2
```

```
# 10. Izbrišite stolpec Staff Skills
df.drop('Staff Skills', axis=1, inplace=True)
```

```
# 11. Shrani DataFrame df v CSV
df.to_csv('university_rank.csv', sep=';', decimal='.')
```

```
# 12. Izpišite tabelo petih univerz z največjim Rank-om
top5 = df.sort_values(by='Rank', ascending=False).head()
print(top5)
```

```
# 13. Izrišite boxplot za stolpec Citations
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Citations'])
plt.show()
```

```
# 14. Izrišite graf za International_Outlook
plt.figure(figsize=(8, 6))
sns.countplot(x='International_Outlook', data=df)
plt.show()
```

```
# 16. Izrišite graf raztrosa za Number_students in Score_Result
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Number_students', y='Score_Result', hue='International_Outlook', palette='coolwarm', data=df)
plt.show()
```

```
# 17. Izrišite graf korelacij (heatmap)
correlation_matrix = df.corr(numeric_only=True) #ad vzamem stevilske vr., ker drugace mam warning :/
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
